```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
perfumeDF = pd.read_csv("perfumes_table.csv")
```

```python
perfumeDF.head(5)
```

```python
perfumeDF['designer'] = perfumeDF['designer'].str.replace(' perfumes and colognes', '')
perfumeDF['title'] = perfumeDF['title'].str.split(" for ",n=1).str[0]
perfumeDF.drop(["rating","reviews","description"], axis = 1, inplace=True)
perfumeDF = perfumeDF[["designer","title","notes","url"]]    

perfumeDF = perfumeDF[perfumeDF['notes'].apply(lambda x: len(x) > 0)]
perfumeDF.head(10)
print(len(perfumeDF))
```

```python
#perfumeDF['notes_not_empty'] = perfumeDF['notes'].apply(lambda x: x != "[]")
perfumeDF = perfumeDF[perfumeDF['notes'].apply(lambda x: x!= "[]")]
#perfumeDF.drop(["notes_not_empty"], axis =1, inplace=True)

# Display the DataFrame with the new column
perfumeDF.tail(20)
print(len(perfumeDF))

perfumeDF = perfumeDF[~perfumeDF.applymap(lambda x: x == '').any(axis=1)]
print(len(perfumeDF))

```

```python
def extract_id(url):
    match = re.search(r'(\d+)(?=\.html)', url)
    return match.group(1) if match else None  # Return the matched number or None if no match

# Apply the function to create a new 'ID' column
perfumeDF['ID'] = perfumeDF['url'].apply(extract_id)
perfumeDF = perfumeDF[["ID","designer","title","notes"]]
perfumeDF['designer'] = perfumeDF['designer'].str.title()
perfumeDF['title'] = perfumeDF['title'].str.title() 

def clean_title(row):
    if row['title'].startswith(" "):
        row['title'] = row['title'][1:] 
    if row['title'].endswith(" "):
        row['title'] = row['title'][:-1] 
    return row

perfumeDF = perfumeDF.apply(clean_title, axis=1)

perfumeDF.tail(20)
```

```python
def remove_designer_substr(row):
    if row['designer'] in row['title']:
        return row['title'].replace(row['designer'], '')
    return row['title']

# Apply the function to each row
perfumeDF['title'] = perfumeDF.apply(remove_designer_substr, axis=1)
perfumeDF.tail(50)
```

```python
gucci_rows = perfumeDF[perfumeDF['designer'] == 'Valentino']

gucci_rows = gucci_rows.apply(clean_title, axis=1)

#guccidict = gucci_rows.set_index('title')['ID'].to_dict()
#print(guccidict)


gucci_rows.head(50)

```

```python
def get_id(df, title, designer):
    # Filter the DataFrame for the given title and designer
    matching_row = df[(df['title'] == title) & (df['designer'] == designer)]
    #print(matching_row)
    # Check if any matching row is found
    if not matching_row.empty:
        # Print the ID of the matching row
        return matching_row['ID'].iloc[0]  # Assuming there is only one match
    else:
        print("No match found for the given title and designer.")

perfumeDF = perfumeDF.apply(clean_title, axis=1)
# Example usage:
get_id(gucci_rows, '1921', 'Gucci')
```

```python
target1 = get_id(perfumeDF, "Black Opium", 'Yves Saint Laurent')
target2 = get_id(perfumeDF, 'Hypnotic Poison', 'Dior')
target3 = get_id(perfumeDF, 'Oud Maracujá', 'Maison Crivelli')
print(target1, target2, target3)
```

```python
def get_notes(df, id_value):
    matching_row = df[df['ID'] == id_value]
  
    if not matching_row.empty:
        retStr = (matching_row['notes'].iloc[0])  # Print the notes of the matching row
        cleaned_str = retStr.replace("[", "").replace("]", "").replace("'", "").replace(",", "").split()
        return cleaned_str
    else:
        print("No match found for the given ID.")

get_notes(perfumeDF, str(83177)) 
```

```python
def shingle(notes, k):
    shingle_list= []
    for i in range(len(notes)-1):
        note_list = []
        for x in range(i,i+k):
            note_list.append(notes[x])
        shingle_list.append(note_list)
        #print(shingle_list)
    return shingle_list

my_list = ['Cardamon', 'Black Pepper', 'Oak', 'Agarwood',"Tea","Vanilla"]
shingle(get_notes(perfumeDF, str(83177)) , 2)
```

```python
def find_jacc_sim(target, current):
    target_shingles = set(shingle(get_notes(perfumeDF, str(target)) , 2))
    current_shingles = set(shingle(get_notes(perfumeDF, str(current)) , 2))
    jacc_simm = len(target_shingles.intersection(current_shingles)) / len(target_shingles.union(current_shingles))
    print( "Jaccard Similarity: " + jacc_simm)
    return jacc_simm

find_jacc_sim(25324, 219)
```
